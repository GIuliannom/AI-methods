### Simple visualization

Testing and learning a few useful and simple visualization methods


```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Auto.csv")
df = pd.DataFrame(data)
```

```python
df = df.dropna()
df #dropping the NaN
```

```python
print('min\n', np.min(df)) #using np.min and np.max to get the range of the data
print()
print('max\n',np.max(df))
```

```python
df.describe() #using describe to see the values that i got are correct
```

```python
df1 = df.drop(df.index[9:85])
df1 #dropping the rows 10 to 85
```

```python
df1.describe() #using describe to find range, mean and standard deviation, easier if going to compare with df
```

```python
import matplotlib.pyplot as plt
```

```python
ax = pd.plotting.scatter_matrix(df,figsize = (10, 10))
plt.show()
```

```python
data = pd.read_csv("CodParasite.txt", delimiter= '\t')
df = pd.DataFrame(data)
df
```

```python
df['Log_int'] = np.log(df.Intensity + 1)
df #Adding one to intensity and taking the log of the column
```

```python
boxplot = df.boxplot(column = ['Log_int'], by =['Sex'], grid = True, figsize=[5,5]) #The boxplot of Log intensity conditional on sex
```

```python
boxplot = df.boxplot(column = ['Log_int'], by =['Stage'], grid = True, figsize=[5,5]) #The boxplot of Log intensity conditional on Stage
```

```python
boxplot = df.boxplot(column = ['Log_int'], by =['Year'], grid = True, figsize=[5,5]) #The boxplot of Log intensity conditional on Year
```

```python
boxplot = df.boxplot(column = ['Log_int'], by =['Area'], grid = True, figsize=[5,5]) #The boxplot of Log intensity conditional on Area
```

```python
data = pd.read_csv("Owls.txt", delimiter= '\t')
df = pd.DataFrame(data)
df
```

```python
boxplot = df.boxplot(column = ['NegPerChick'], by =['FoodTreatment'], grid = True, figsize=[5,5]) #The boxplot of negperchick conditional on FoodTreatment
```

```python
boxplot = df.boxplot(column = ['NegPerChick'], by =['FoodTreatment', 'SexParent'], grid = True, figsize=[10,10])#The boxplot of negperchick conditional on sexparent and foodtreatment
```
